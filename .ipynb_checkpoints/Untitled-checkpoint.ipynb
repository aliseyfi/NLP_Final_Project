```python
# Find the tweets in US
def find_us_tweets(data):
    e_p_li = pd.DataFrame(columns=['text', 'date', 'state'])
    for text, date, loc in zip(data['text'], data['created_at'], data['place_full_name']):
        state = loc.split(',')[0]
        if state in states_li['State'].values:
            abbre = states_li[states_li['State']==state]['Abbreviation'].values[0]
            e_p_li = e_p_li.append({'text': text, 'date': date, 'state':abbre}, ignore_index=True)
    return e_p_li
```

Remove URLs and space in mentions, hashtags.

```python
# Remove URLs
def preprocessing(tweets_all):
    url_pattern = re.compile('http[s]?://(?:[a-zA-Z]|[0-9]|[$-_@.&+]|[!*\(\),]|(?:%[0-9a-fA-F][0-9a-fA-F]))+')
    pic_pattern = re.compile('pic\.twitter\.com/.{10}')
    tweets_all['text'] = tweets_all['text'].apply(lambda buf: url_pattern.sub('',buf))
    tweets_all['text'] = tweets_all['text'].apply(lambda buf: pic_pattern.sub('',buf))

    # removing space in mentions and hashtags
    tweets_all['text'] = tweets_all['text'].apply(lambda buf: buf.replace('@ ', '@'))
    tweets_all['text'] = tweets_all['text'].apply(lambda buf: buf.replace('# ', '#'))
    return tweets_all
```

### Tokenization

I use TweetsTokenizer here to tokenize the tweets.

```python
porter_stemmer = PorterStemmer()
treebank_tokenizer = TreebankWordTokenizer()
tweet_tokenizer = TweetTokenizer()
sentence_tokenizer = nltk.data.load('tokenizers/punkt/english.pickle')
lemmatizer = WordNetLemmatizer()
stop_words = nltk.corpus.stopwords.words('english')+["http","https","www", "com"]

# Tokenize tweets
def tokenize(text, Tweets_token = True):
    punkt_sentences = sentence_tokenizer.tokenize(text)
    
    # Use tweets tokenization or tree bank tokenization
    if not Tweets_token:
        sentences_words = [treebank_tokenizer.tokenize(sentence) for sentence in punkt_sentences]
    else:
        sentences_words = [tweet_tokenizer.tokenize(sentence) for sentence in punkt_sentences]
        
    all_tokens = [lemmatizer.lemmatize(word.lower()) for sentence in sentences_words for word in sentence if word.isalpha() and word.lower() not in stop_words]
    bigrams = nltk.bigrams(all_tokens)
    return all_tokens # the later function assumes you are returning a list of terms
```

I calculate the sum of intensity of each emotions and save as a list of dictionary in form: [{user_id: [{emotion: intensity},[token1, token2, ...]]}].

Meanwhile, save the emotion word used in the calculation to draw a wrodcloud.

```python
# Find emotion for each tweets
def Find_emotion(token_dict, emotion_dict):
    eDict = []
    emotion_word = [] 
    for userid_tokenlist in token_dict:
        for userid, token_li in userid_tokenlist.items():
            subE = {'anticipation': 0, 'fear': 0, 'anger': 0, 'trust': 0, 'surprise': 0, 'sadness': 0, 'joy': 0, 'disgust': 0}
            for token in token_li:
                if token in emotion_dict.keys():
                    emotion = emotion_dict[token][0]
                    intensity = emotion_dict[token][1]
                    subE[emotion] += intensity
                    emotion_word.append(token)
            eDict.append({userid: [subE, token_li]})
    return eDict, emotion_word
```

Calculate the overall intensity for each emotion.

```python
# Calculate sum of intensity for each emotion
def find_e_all(emotion_per):
    E_all = {'anticipation': 0, 'fear': 0, 'anger': 0, 'trust': 0, 'surprise': 0, 'sadness': 0, 'joy': 0, 'disgust': 0}
    for i in emotion_per:
        for emotion in i.values():
            for k,v in emotion[0].items():
                E_all[k] += v
    return E_all
```

Use a simply rule based system to label the tweets data - simply choose the highest intensity as the emotion label of the tweets.

```python
# Find max intensity
def find_max_intensity(Edict):
    Rdict = {v:k for k,v in Edict.items()}
    emotion = max(list(Rdict.keys()))
    return Rdict[emotion]

# Label tweets by max intensity emotion
def label_text(Emotion_dict):
    new_dict = []
    for i in Emotion_dict:
        for k, v in i.items():
            new_dict.append({k:[v[1], find_max_intensity(v[0])]})
    return new_dict

# Label all tweets
def get_labeled_data(emotion_per):
    train = pd.DataFrame(columns=['text','emotion'])
    for i in label_text(emotion_per):
        for j in i.values():
            text = ' '.join(j[0])
            lrow = pd.DataFrame({'text':text, 'emotion': j[1]}, index=[1])
            train = train.append(lrow, ignore_index=True)
    return train
```

Define a funciton to train model with specific number of features.

```python
def train_model(m_dict, k, X_train, y_train, X_test, y_test):
    model_df = pd.DataFrame(columns=['model', 'accuracy'])
    
    for name, model in m_dict.items():
        pipeline = Pipeline([('vect', vectorizer),
                             ('chi',  SelectKBest(chi2, k=k)),
                             ('clf', model)])
        model = pipeline.fit(X_train, y_train)
        ytest = np.array(y_test)
        score= model.score(X_test, ytest)
        model_df = model_df.append({'model': name, 'accuracy': score}, ignore_index=True)
        print(classification_report(ytest, model.predict(X_test)))
        print(confusion_matrix(ytest, model.predict(X_test)))
    return model_df
```

### Creating the dictionary of the parameter grids
- In the dictionary:
    - the key is the acronym of the model
    - the value is the parameter grid of the model

```python
param_grids = {}
```

#### The parameter grid for logistic regression
The hyperparameters we want to fine-tune are:
- C
- tol

See details of the meaning of the hyperparametes in [sklearn.linear_model.LogisticRegression](https://scikit-learn.org/stable/modules/generated/sklearn.linear_model.LogisticRegression.html)

```python
# The grids for C
C_grids = [10 ** i for i in range(-2, 3)]

# The grids for tol
tol_grids = [10 ** i for i in range(-6, -1)]

# Update param_grids
param_grids['lr'] = [{'model__C': C_grids,
                      'model__tol': tol_grids}]
```

#### The parameter grid for decision tree
The hyperparameters we want to fine-tune are:

### Hyperparameter tuning
The code below shows how to fine-tune the hyperparameters of the models above (using sklearn GridSearchCV).

```python
# Make directory
directory = os.path.dirname('./cv_results/')
if not os.path.exists(directory):
    os.makedirs(directory)
```

Train models by using grid search and save the model as pickle in the disk. 

```python
from sklearn.model_selection import GridSearchCV

# The list of [best_score_, best_params_, best_estimator_] obtained by GridSearchCV
best_score_param_estimator_gs = []

for acronym in pipes.keys():
    # GridSearchCV
    gs = GridSearchCV(estimator=pipes[acronym],
                      param_grid=param_grids[acronym],
                      scoring='f1_micro',
                      n_jobs=2,
                      cv=ps,
                      return_train_score=True)
        
    # Fit the pipeline
    gs = gs.fit(X_train_valid, y_train_valid)
    
    print('success!')
    print(gs.best_score_)
    
    # Save model
    save_Instance(gs, '{model}_half_features'.format(model=acronym))
        
    # Update best_score_param_estimator_gs
    best_score_param_estimator_gs.append([gs.best_score_, gs.best_params_, gs.best_estimator_])
    
    # Sort cv_results in ascending order of 'rank_test_score' and 'std_test_score'
    cv_results = pd.DataFrame.from_dict(gs.cv_results_).sort_values(by=['rank_test_score', 'std_test_score'])
    
    # Get the important columns in cv_results
    important_columns = ['rank_test_score',
                         'mean_test_score', 
                         'std_test_score', 
                         'mean_train_score', 
                         'std_train_score',
                         'mean_fit_time', 
                         'std_fit_time',                        
                         'mean_score_time', 
                         'std_score_time']
    
    # Move the important columns ahead
    cv_results = cv_results[important_columns + sorted(list(set(cv_results.columns) - set(important_columns)))]

    # Write cv_results file
    cv_results.to_csv(path_or_buf='./cv_results/' + acronym + '.csv', index=False)
```

```python
q = pd.DataFrame(columns=['model', '2000_features', '5000_features', 'half_features'])
for acronym in pipes.keys():
    subD = {'model': acronym}
    for number_of_features in ['2000_features', '5000_features', 'half_features']:
        mn = '{model}_{number}'.format(model=acronym, number=number_of_features)
        score = load_Instance('save_pickle/{}'.format(mn)).best_score_
        subD[number_of_features] = score
    q = q.append(subD, ignore_index=True)

q
```
